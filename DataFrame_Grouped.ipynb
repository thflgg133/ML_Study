```python
# 전복(abalone) 성별에 따라 groupby함수를 통해 집계를 하겠습니다.
# ex) DataFrame[집계 변수].groupby(DataFrame[집계 대상])
grouped = abalone_df['whole_weight'].groupby(abalone_df['sex'])
```

```python
# 그룹변수를 2개 이상 선택해서 총계처리

abalone_df.groupby(['sex', 'length_bool']).mean() # => EDA (EXPLORATORY DATA ANALYSIS)
```

```python
# 데이터 프레임의 꽃
# apply 함수 매우중요
본인이 원하는 행과 열에 연산 혹은 function을 적용할 수 있다.
axis = 0 열
axis = 1 행
```

```python
# abalone_df[['diameter']].mean()
# 열 기준

abalone_df[['diameter']].apply(np.average, axis = 0)
```

```python
# 행 기준

abalone_df[['diameter','whole_weight']].apply(np.average, axis = 1)
```

```python
# 사용자 함수를 통한 집계 ★★★★★

import math
def avg_ceil(x,y,z):
  return math.ceil(x+y+z) # ceil == 올림 ex) 3.7 -> 4

abalone_df[['diameter', 'height','whole_weight']].apply(lambda x :avg_ceil(x[0],x[1],x[2]),axis=1)
```

```python
def sum_data(x,y,z): # 사용자 정의 함수!! 정말 중요
  if x+y+z > 1:
    return True
  
  else:
    return False

answer = abalone_df[['diameter','height','whole_weight']].apply(lambda x : sum_data(x[0],x[1],x[2]),axis=1)
abalone_df['answer'] = answer
abalone_df
```
