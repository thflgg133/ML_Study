```python
#슬라이싱 기능을 통해 여러 행 가져오기
df.iloc[[2,3]] # 슬라이싱 X
```

```python
df.iloc[2:4] #슬라이싱
```

```python
# iloc vs loc 차이
iloc는 인덱스와 컬럼을 리스트 배열로 선택하는 것
loc는 인덱스와 컬럼을 문자로 선택하는 것
```

```python
# 조건 필터링 가져오기
# 30대 이상만
df[df['age']  >= 30]
```

```python
## job이 intern인 사람 가져오기
df[df['job'] == 'intern']
```

```python
## 조건 여러개
## 30대 이상 40대 이하 (and function)
df[(df['age']>=30) & (df['age'] <=40)]
```

```python
 ## 조건 여러개 30대미만 혹은 40대 초과 (or function)
 df[(df['age'] < 30) | (df['age'] > 40)]
```

```python
# in을 통한 포함 조건 걸기
#df[df['job'] in ['student','manager']]
df[df['job'].apply(lambda x : x in ['student','manager'])]
```

```python
df['job'].apply(lambda x : x in ['student','manager'])
```
