```python
# - DataFrame 행, 열 삭제
df = pd.DataFrame(np.arange(12).reshape(3,4), # reshape = 3행, 4열로 만들어준다는 뜻
                  columns=['A','B','C', 'D'])
df
```

```python
df.drop(['B','C'], axis=1, inplace=True) #axis = 0 or 1 // 0 = row , 1 = column
```

```python
df.drop([0,1], axis =0) #axis Defalut 값 == 0
```

```python
df.loc[1,'C'] ='육'
df
```

```python
df.iloc[1,2] = 6
df
```

```python
# 슬라이싱을 기능을 활용하면 내가 원하는 범위까지 잘 수정 가능
df.iloc[:,1] = "끝!"
df 
```
